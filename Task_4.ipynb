```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '/content/adult.csv'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'educational-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
df = pd.read_csv(url, names=columns, na_values=' ?', skiprows=1)
```

```python
df = df.dropna()
```

```python
print("Original dataset shape:", df.shape)
print(df.head())
```

```python
num_features = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
cat_ordinal = ['education']
cat_nominal = ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
target = 'income'
```

```python
# Label Encoding for ordinal
le = LabelEncoder()
df[cat_ordinal] = df[cat_ordinal].apply(le.fit_transform)
```

```python
# One-Hot Encoding for nominal
df = pd.get_dummies(df, columns=cat_nominal, drop_first=True)
```

```python
scaler = StandardScaler()
df[num_features] = scaler.fit_transform(df[num_features])

print("Numerical features scaled successfully.")
print(df[num_features].head())
```

```python
# Scaling numerical features
scaler = StandardScaler()
df[num_features] = scaler.fit_transform(df[num_features])
```

```python
# Before/after scaling comparison
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, feat in enumerate(num_features):
    ax = axes[i//3, i%3]
    df_original = pd.read_csv(url, names=columns, na_values=' ?').dropna()[feat]  # Reload original for comparison
    sns.histplot(df_original, kde=True, ax=ax, color='blue', label='Before')
    sns.histplot(df[feat], kde=True, ax=ax, color='red', label='After')
    ax.legend()
plt.tight_layout()
plt.show()
```
